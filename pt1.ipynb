```python
import numpy as np
import plotly.graph_objs as go
```

```python
x = np.random.normal(loc=100, scale=50, size=500)
y = np.random.normal(loc=200, scale=10, size=500)
y1 = np.random.normal(loc=120, scale=50, size=500)

z = 3*x + 4*y
z1 = 3*x*x + 4*y
z2 = 3*x*x + 4*y*y
```

```python
# Create the scatter plot
fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z, mode='markers')])

# Update layout for better visualization
fig.update_layout(scene=dict(
                    xaxis_title='X Axis',
                    yaxis_title='Y Axis',
                    zaxis_title='Z Axis'),
                  title='3D Scatter Plot')

# Show plot
fig.show()

# Create the scatter plot
fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z1, mode='markers')])

# Update layout for better visualization
fig.update_layout(scene=dict(
                    xaxis_title='X Axis',
                    yaxis_title='Y Axis',
                    zaxis_title='Z Axis'),
                  title='3D Scatter Plot')

# Show plot
fig.show()


# Create the scatter plot
fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z2, mode='markers')])

# Update layout for better visualization
fig.update_layout(scene=dict(
                    xaxis_title='X Axis',
                    yaxis_title='Y Axis',
                    zaxis_title='Z Axis'),
                  title='3D Scatter Plot')

# Show plot
fig.show()
```

```python
import numpy as np
import plotly.figure_factory as ff



# Create the 2D density plot
fig = ff.create_2d_density(y1, y, colorscale='Viridis')

# Update layout for better visualization
fig.update_layout(
    scene=dict(
        xaxis_title='Y1 Axis',
        yaxis_title='Y Axis',
        zaxis_title='Density'
    ),
    title='3D Joint Density Plot'
)

# Show plot
fig.show()
```
